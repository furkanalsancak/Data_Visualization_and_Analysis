```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [1, 50]
y = [value * 3 for value in x]

plt.plot(x,y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line')

plt.show()
```

```python
x1 = [1,2,3]
y1 = [2,4,1]
locs = [1.0,1.5,2.0,2.5,3.0]
# labels = ['A','B','C','D','E']

plt.plot(x1,y1)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample Graph!')
# plt.xticks(locs, labels)
plt.xticks(locs)
plt.show()
```
